```python
def recent_form(df, team, idx, n=5):
    past = df.iloc[:idx]
    matches = past[(past["team1"] == team) | (past["team2"] == team)].tail(n)    wins = (matches["winner"] == team).sum()
    return wins / n if n > 0 else 0
```
